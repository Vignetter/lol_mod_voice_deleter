```python
from pathlib import Path
```

```python
ROOT = "D:/User/Saved Games/LOL/JP/"
ROOT += "assets/sounds/wwise2016/vo/zh_tw/"
TXT = r'D:/User/Saved Games/LOL/mod/--voice delete--.txt'
```

```python
def get_champion_list():
    champion_list = []
    with open(TXT) as file:
        for line in file.readlines():
            champion_list.append(line.strip('\n'))
    return sorted(champion_list)
```

```python
def delete_all_voice(champions):
    delete_announcer()
    print("-"*8)
    for champion in champions:
        dir = Path(ROOT, 'characters', champion, 'skins', 'base')
        print(f'Dealing with {champion} voice')
        if dir.exists():
            print(f"{champion}'s voice exist")
            for file in dir.iterdir(): 
                file.unlink()
            dir.rmdir()
            print(f"{champion}'s voice have been deleted.")
        else:
            print(f"{champion}'s voice is not exist.")
        if champion == 'darius': delete_extra_darius()
        if champion == 'quinn': delete_extra_quinn()
        if champion == 'ziggs': delete_extra_ziggs()
        print("-"*8)
```

```python
def delete_extra_quinn():
    champion = 'quinn'
    dir = Path(ROOT, 'characters', champion, 'skins', 'skin05')
    if dir.exists():
        print(f"{champion}'s extra voice exist.")
        for file in dir.iterdir():
            file.unlink()
        dir.rmdir()
        print(f"{champion}'s extra voice have been deleted.")
    else:
        print(f"{champion}'s extra voice is not exist.")
```

```python
def delete_extra_darius():
    champion = 'darius'
    dir = Path(ROOT, 'characters', champion, 'skins', 'skin15')
    if dir.exists():
        print(f"{champion}'s extra voice exist.")
        for file in dir.iterdir():
            file.unlink()
        dir.rmdir()
        print(f"{champion}'s extra voice have been deleted.")
    else:
        print(f"{champion}'s extra voice is not exist.")
```

```python
def delete_extra_ziggs():
    champion = 'ziggs'
    dir = Path(ROOT, 'characters', champion, 'skins', 'skin06')
    if dir.exists():
        print(f"{champion}'s extra voice exist.")
        for file in dir.iterdir():
            file.unlink()
        dir.rmdir()
        print(f"{champion}'s extra voice have been deleted.")
    else:
        print(f"{champion}'s extra voice is not exist.")
```

```python
def delete_announcer():
    file_list = ['announcer_global_female1_vo_audio.wpk', 'announcer_global_female1_vo_events.bnk']
    print('Dealing with [announcer] voice')
    i = 1
    for file_name in file_list:
        file = Path(ROOT, 'shared', file_name)
        if file.exists():
            file.unlink()
            print(f"announcer voice{i} have been deleted.")
        else:
            print(f"announcer voice{i} is not exist.")
        i += 1
```

```python
# main execution
champions = get_champion_list()
delete_all_voice(champions)
```
